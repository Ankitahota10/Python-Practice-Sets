# Filtering & Sorting

```python
# clean the item_price column and transform it in a float
chipo['item_price'] = [i[1:] for i in chipo.item_price]
chipo[['item_price', 'item_name']]
```

```python
# delete the duplicates in item_name and quantity
no_dup = chipo.drop_duplicates(['item_name','quantity','choice_description'])
no_dup
```

```python
#select only the products with quantity equals to 1
q1 = chipo[chipo['quantity']==1]
q1
```

```python
# select only the item_name and item_price columns
cols = chipo[['item_name','item_price']]
cols
#sort the values from the most to less expensive
cols.sort_values(by = 'item_price',ascending = False)
#OR : cols['item_price'].sort_values(ascending = False)
```

```python
#Sort by the name of the item
cols['choice_description'] = chipo['choice_description']
cols['item_name'].sort_values(ascending= True).unique()
```

```python
#What was the quantity of the most expensive item ordered?
#chipo['item_price'].sort_values(ascending = False).head()
#OR
chipo.sort_values(by = "item_price", ascending = False).head(5)
```
