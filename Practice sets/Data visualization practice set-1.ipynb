```python
'''Plot Total Sales Per Month for Year 2011. How the total sales haveincreased over months in Year 2011. 
Which month has lowest Sales?'''
mart= pd.read_csv('C:/Users/ANKITA HOTA/Documents/Python/Edureka case studies/Module4/BigMartSalesData.csv', delimiter=',')
mart.head()
```

```python
mart_sales= mart[mart['Year']==2011].groupby('Month').agg('sum').filter(['Amount']).sort_values('Amount')
mart_sales
```

```python
x1= mart[mart['Year']==2011].groupby('Month')
y1= mart[mart['Year']==2011].groupby('Month').sum()['Amount']
plt.plot(y1.index,y1.values, 'm*-')
plt.xlabel('Months in 2011')
plt.ylabel('Total sales')
plt.title('Total sales per months in 2011')
```

```python
'''Plot Total Sales Per Month for Year 2011 as Bar Chart.  Is Bar Chart Better to visualize than Simple Plot?'''
x1= mart[mart['Year']==2011].groupby('Month').sum()['Amount']
plt.bar(x1.index,x1.values,color='purple')
```

```python
#Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?
country= mart[mart['Year']==2011].groupby('Country')
max_sales= country.sum()['Amount']
plt.figure(figsize=(14,14))
plt.pie(max_sales.values, labels=max_sales.index,autopct='%1.1f%%')
```

```python
'''Plot Scatter Plot for the invoice amounts and see the concentration of amount.
In which range most of the invoice amounts are concentrated'''
#mart.head()
invoice= mart[mart['Year']==2011].groupby('InvoiceNo').sum()['Amount']
plt.scatter(invoice.index, invoice.values)
plt.xlim(0,250)
plt.ylim(0,250)
```
